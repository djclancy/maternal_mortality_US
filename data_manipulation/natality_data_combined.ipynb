```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

First, we'll make the Natality data by year into a data frame.

```python
path_to_file = '../data/natality_data/Natality2010-2023.txt'
path_to_file_csv = '../data/natality_data/Natality2010-2023.csv'

txt_line_to_csv_line = lambda x : x.replace("\t", ',')
write_file = open(path_to_file_csv, 'w')
f = open(path_to_file)
line_number = -1
for line in f:
    line_number+=1
    if "---" in line:
        break
    if  not ("Total" in line):
        csv_line = txt_line_to_csv_line(line)
        write_file.write(csv_line)
natality_df = pd.read_csv(path_to_file_csv)
natality_df = natality_df.rename(columns={'Census Division': 'Division'})
natality_df.Division = natality_df.Division.apply(lambda x: x[12:])
natality_df  = natality_df[['Division','Year','Births']]
natality_df.head()
```
